Plotting In An Object Oriented Way

```python
%matplotlib inline
fig,ax=plt.subplots(figsize=(5,10))

x=[1,2,3]
y=[11,22,33]

ax.plot(x,y)

ax.set(title="Simple Plot",
        xlabel="X-Plot",
        ylabel="Y-Plot")

fig.savefig("plotSaved.png")
```

# lets make a bar plot

```python
animes={"AOT":30,"Naruto":20,"FMAB":10}
fig,ax=plt.subplots()
ax.bar(animes.keys(),animes.values())
```

```python
animes={"AOT":30,"Naruto":20,"FMAB":10}
fig,ax=plt.subplots()
ax.barh(list(animes.keys()),list(animes.values()))
```

# Histograms

```python
x=np.random.randn(1000)
fig,ax=plt.subplots()
ax.hist(x)
```

# subplots

```python
fig,((ax1,ax2),(ax3,ax4))=plt.subplots(nrows=2,ncols=2,figsize=(10,5))

ax1.plot(x,x/2)
ax2.scatter(np.random.random(20),np.random.random(20))
ax3.bar(animes.keys(),animes.values())
ax4.hist(x)
```

# SubPlots with another syntax

```python
fig,ax=plt.subplots(nrows=2,ncols=2,figsize=(10,5))

ax[0,0].plot(x,x/2)
ax[0,1].scatter(np.random.random(20),np.random.random(20))
ax[1,0].bar(animes.keys(),animes.values())
ax[1,1].hist(x)
```
